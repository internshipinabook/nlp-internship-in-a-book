## EfficientNet-B0: Compound Scaling

> 🧠 **What Will This Output?**
> EfficientNet-B0 has 5.3M params vs ResNet-18's 11.7M. Predict: will EfficientNet converge faster or slower? What property of compound scaling might make it more or less stable than a residual network?

```python
def build_efficientnet_b0(num_classes=4,pretrained=True):
    weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1 if pretrained else None
    m=models.efficientnet_b0(weights=weights)
    m.classifier=nn.Sequential(nn.Dropout(0.3),nn.Linear(1280,num_classes))
    return m

model_eff=build_efficientnet_b0().to(DEVICE)
n_eff=sum(p.numel() for p in model_eff.parameters())
print(f"EfficientNet-B0: {n_eff:,} params")
print(f"ResNet-18:       11,681,476 params")
print(f"Ratio:           {11681476/n_eff:.1f}x larger")
```

```python
# Lower LR than ResNet — EfficientNet is more sensitive
print("Training EfficientNet-B0 (20 epochs, differential LR)...")
history_eff,best_eff,ep_eff=run_training(
    model_eff,df,n_epochs=20,lr_head=5e-4,lr_backbone=5e-6,
    save_path="models/efficientnet_best.pth",verbose=True
)
print(f"EfficientNet best val F1: {best_eff:.3f}")
```
